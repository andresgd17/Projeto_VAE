## Jupyter-notebook for reading data file

```python
import numpy as np
import h5py
```

```python
hf = h5py.File('img_p+Fe_MST+SST.hdf5', 'r')
```

```python
hf.keys()
```

```python
MST = hf.get('MST')
MST.keys()
```

```python
X_MONO_MST   = np.array(MST['X_MONO'])
X_STEREO_MST = np.array(MST['X_STEREO'])
ENERGY_MST   = np.array(MST['ENERGY'])
XMAX_MST     = np.array(MST['XMAX'])
LABEL_MST    = np.array(MST['LABEL'])
Y_MST        = np.array(MST['Y'])
```

```python
X_STEREO_MST.shape, X_MONO_MST.shape, ENERGY_MST.shape, XMAX_MST.shape, Y_MST.shape, LABEL_MST.shape
```

 A primeira componente dos arrays contem os eventos de ferro, e a segunda componente os eventos de proton

```python
X_STEREO_MST[0].shape, X_MONO_MST[0].shape, ENERGY_MST[0].shape, XMAX_MST[0].shape, Y_MST[0].shape, LABEL_MST[0]
```

```python
X_STEREO_MST[1].shape, X_MONO_MST[1].shape, ENERGY_MST[1].shape, XMAX_MST[1].shape, Y_MST[1].shape, LABEL_MST[1]
```

Cada classe contem 1900 imagens com sua correspondente energia. Os Y sÃ£o vetores para representar cada classe: [1 0] para ferro, e [0 1] para proton. Segue o loop para entender melhor.

```python
for label, array_img_stereo, array_img_mono, array_energy, array_xmax, array_y in zip(LABEL_MST, X_STEREO_MST, X_MONO_MST, ENERGY_MST, XMAX_MST, Y_MST):
    print('particle: ', label, ' --> array de 1900 imagens de 112x112 pixels', array_img_stereo.shape, array_energy.shape)
    for img_stereo, img_mono, energy, xmax, y in zip(array_img_stereo, array_img_mono, array_energy, array_xmax, array_y):
        print('particle: ', label, 'image shape:', img_mono.shape, ' Energy =', energy, '  Xmax =', xmax,'  y:', y)
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

fig = plt.figure(figsize=(20,20))
for i in range(4):
    ax = fig.add_subplot(2, 2, i + 1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(X_STEREO_MST[0][i]))
    xlabel = "{0}".format(LABEL_MST[0])
    xtitle = "E = %i TeV -- $X_{max}$ = %i g/cm$^{2}$" %(ENERGY_MST[0][i], XMAX_MST[0][i])
    # Show the classes as the label on the x-axis.
    ax.set_xlabel(xlabel,fontsize=31)
    ax.set_title(xtitle, fontsize=31)
    
```
